What is the number of distinct countries present in the database?

```python
--num_distinct_countries
select 
	count(distinct country_name) total_distinct_countries
from
	international_debt;
```

What are the distinct debt indicators?

```python
--distinct_debt_indicators
select distinct
	indicator_code total_distinct_countries
from
	international_debt;
```

What is the total amount of debt owed by all the countries present in the table, in millions, rounded to two decimals?

```python
--total_debt
select 
	round(sum(debt) / 1000000, 2)  total_debt 
from
	international_debt;
```

What country has the highest amount of debt?

```python
--highest_debt_country
select
	country_name,
	round(sum(debt), 2) total_debt
from
	international_debt
group by country_name
order by total_debt desc
limit 1;
```

```python
--avg_debt_per_indicator
select 
	indicator_code debt_indicator,
	indicator_name,
	avg(debt) average_debt
from 
	international_debt
group by debt_indicator, indicator_name
order by average_debt desc;
```

What country has the highest amount of principal repayments in the "DT.AMT.DLXF.CD" category?

```python
--highest_principal_repayment
select
	country_name,
	indicator_name
from
	international_debt
where debt = (select max(debt) 
			  from international_debt 
			  where indicator_code='DT.AMT.DLXF.CD');
```
